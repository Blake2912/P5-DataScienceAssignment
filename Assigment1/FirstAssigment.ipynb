```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
from csv import writer
from csv import reader
default_text = 'NaN'
# Open the input_file in read mode and output_file in write mode
with open('data.csv', 'r') as read_obj, open('data1.csv', 'w', newline='') as write_obj:
    # Create a csv.reader object from the input file object
    csv_reader = reader(read_obj)
    # Create a csv.writer object from the output file object
    csv_writer = writer(write_obj)
    # Read each row of the input csv file as list
    for row in csv_reader:
        # Append the default text in the row / list
        row.append(default_text)
        # Add the updated row / list to the output file
        csv_writer.writerow(row)
output_data=pd.read_csv('data.csv')
output_data.head()
```

```python
data = pd.read_csv('data1.csv')
```

```python
print(data.shape)
print(data.head(5))
```

```python
# To display the information of the data
data.info()
```

```python
data.isnull().sum()
```

```python
# Here we are using a lambda function to check if the diagnosis parameter is M or not, if the value of X=='M'
# Then set the diagnosis to 0 else 1
data['diagnosis'] = data['diagnosis'].apply(lambda x: 0 if x=='M' else 1) 
data.head()
```

```python
y = data['diagnosis']
X = data.drop(['id','diagnosis','NaN'], axis = 1)
"""
The drop function is used to drop rows and columns in the dataset and this is a function from the Pandas library,
the args to be passed to the function are 'label' i.e. the rows/columns to be dropped and axis=1 states that we are
dropping columns and not rows and it defaults to 0 which is rows.
"""
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
# For testing purposes do not Uncomment
# print(X_train)
# print(y_train)
```

```python
#Linear Regression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Predicting Model
pred = model.predict(X_test)
y_pred1 = [ 0 if x < 0.5 else 1 for x in pred]
y_pred1
```

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
# Prediction
y_pred2 = log_model.predict(X_test)
y_pred2
```

```python
#Confusion Matrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cm1= confusion_matrix(y_test,y_pred1) # This is from the linear regression
cm2= confusion_matrix(y_test,y_pred2) # y_pred2 This is for Logistic Regression
print("Confusin Matrix: ")
print(cm1)
print(cm2)
```

```python
#Linear Regression Confusion Matrix
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm1, display_labels =[False,True])
cm_display.plot()
plt.show()
```

```python
#Print Accuracy of the model
print("Accuracy:",metrics.accuracy_score(y_test, y_pred1))
```

```python
#ROC Curve -Linear
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred1)
roc_auc = auc(false_positive_rate, true_positive_rate)
print(roc_auc)
plt.plot(false_positive_rate,true_positive_rate)
```

```python
#ROC Curve Logistic 
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred2)
roc_auc = auc(false_positive_rate, true_positive_rate)
print(roc_auc)
plt.plot(false_positive_rate,true_positive_rate)
```
